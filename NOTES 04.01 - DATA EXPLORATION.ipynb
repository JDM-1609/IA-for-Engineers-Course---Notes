### correlations

Mostramos en una matriz de correlaciones entre cada una de las variables con un cÃ³digo de color.
La correlaciÃ³n es una medida estadÃ­stica que indica la relaciÃ³n y la fuerza de la asociaciÃ³n entre dos variables.
Mide cÃ³mo cambian las variables juntas, seÃ±alando si el cambio en una se asocia con un aumento (correlaciÃ³n positiva) o una disminuciÃ³n (correlaciÃ³n negativa) en la otra.

```python
#correlation matrix
corrmat = d.select_dtypes(include=np.number).corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```

## Inspect categorical variables
Las variables categoricas son las que no tiene asignado un nÃºmero sino, generalmente, una cadena de caracteres.

```python
## KEEPOUTPUT
ccols = [i for i in d.columns if not i in d._get_numeric_data()]
print (ccols)
```

```python
## KEEPOUTPUT
for c in ccols:
    print ("%10s"%c, np.unique(d[c].dropna()))
```

```python
## KEEPOUTPUT
c="GarageType"
d[c].value_counts()
```

```python
## KEEPOUTPUT
plt.figure(figsize=(20,8))
for i,c in enumerate(["ExterQual", "HouseStyle", "LandSlope", "Alley"]):
    plt.subplot(2,4,i+1)
    k=d[[c,"SalePrice"]].dropna()
    for v in d[c].dropna().unique():
        sns.histplot(k.SalePrice[k[c]==v], label=v, kde=True);
        plt.title(c)
    plt.yticks([])
    plt.legend()
    plt.subplot(2,4,i+5)
    vc = k[c].value_counts()
    sns.barplot(x=vc.index, y=vc.values)
    plt.xticks(range(len(vc)), vc.index, rotation="vertical")
```
