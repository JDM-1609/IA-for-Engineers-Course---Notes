**NOTE**: Specify your PNL rounded to **TWO** decimal places

```python
s = ((np.random.normal(size=21)*5+.5).cumsum()+100).round(2)
s
```

```python
# Calcular predicción perfecta base
perfect_pred = (s[1:] > s[:-1]).astype(int)
n_predictions = len(perfect_pred)

# Inicializar arrays
pset = np.zeros((9, n_predictions), dtype=int)
pnl = np.zeros(9)

# Para cada nivel de accuracy (10% a 90%)
for i, target_acc in enumerate(np.linspace(0.1, 0.9, 9)):
    # Crear predicciones con accuracy específico
    pred = perfect_pred.copy()
    n_correct_desired = int(target_acc * n_predictions)
    n_current_correct = n_predictions

    # Degradar predicciones hasta alcanzar accuracy deseado
    indices = np.random.permutation(n_predictions)
    for idx in indices:
        if n_current_correct > n_correct_desired:
            if pred[idx] == perfect_pred[idx]:
                pred[idx] = 1 - pred[idx]
                n_current_correct -= 1
        else:
            break

    pset[i] = pred

    # Calcular PNL para estas predicciones
    direction = (pred - 0.5) * 2
    price_change = s[1:] - s[:-1]
    commission = s[:-1] * 0.01
    total_pnl = np.sum((direction * price_change) - commission)
    pnl[i] = np.round(total_pnl, 2)

# Asegurar que pset sea un numpy array
pset = np.array(pset)
```
